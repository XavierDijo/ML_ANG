```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv ('ex1data1.txt', sep=",", header=None)
```

```python
X = df.values[:,0:1] 
y = df.values[:,1:2]
```

```python
X = np.insert(X, 0, 1, axis=1) #Add column "1"
```

```python
# 
```

```python
def computeCost(X,y,theta):
    m=len(y)
    return np.sum(np.square(X.dot(theta) - y))/(2*m)
```

```python
def gradientDescent(X, y, theta, alpha, num_iters) :
    m=len(y)
    J_history=[]
    
    for i in range(num_iters):
        theta = theta - (alpha * 1/m) * np.dot(X.transpose(),(X.dot(theta) -y))
        J_history.append(computeCost(X,y,theta))
    
    return theta, J_history

```

```python
theta = np.array([[0,],[0,]])
theta,J_history = gradientDescent(X,y,theta,0.01,1500)
```

```python
import matplotlib.pyplot as plt
plt.plot(J_history)
plt.xlabel("Iteration")
plt.ylabel("Cost")
plt.title("Gradient Descent")
```

# Normal Equations

```python
from numpy.linalg import inv
```

```python
# normal equation  
# theta_best = (X.T * X)^(-1) * X.T * y  
```

```python
X_transpose = X.T  
theta_best = inv(X_transpose.dot(X)).dot(X_transpose).dot(y)  
```

```python
theta_best
```

```python
plt.scatter(df.values[:,0:1],y)



x_value=[x for x in range(25)]
y_value=[y*theta_best[1]+theta_best[0] for y in x_value]
plt.plot(x_value,y_value,color="g", label ="normal equation")

x_value=[x for x in range(25)]
y_value=[y*theta[1]+theta[0] for y in x_value]
plt.plot(x_value,y_value,color="r", label="GD")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

plt.show()
```
